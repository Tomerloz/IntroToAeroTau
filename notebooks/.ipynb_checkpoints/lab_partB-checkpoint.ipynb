# Vortex Panel Method

The **vortex panel method** models potential flow around a solid body (like an airfoil) by distributing *vortex strength* along the body’s surface. The body is discretized into straight panels, and at each panel we solve for the vortex strength \(\gamma\) such that the boundary condition (no flow normal to the surface) is satisfied at the panel center (“collocation point”).

For each panel, the *sum* of the normal component of the freestream and the induced velocity from all vortex panels is set to zero. This coupling leads to a linear system of equations, where the unknowns are the strengths of each vortex panel.

In this notebook, we:
- **Read the geometry** of a **NACA0012 airfoil** from file.
- **Discretize** the surface into panels.
- **Assemble and solve** the linear system for the vortex strengths (\(\gamma\)).
- **Compute the pressure coefficient \(C_p\)** on the airfoil surface.
- **Visualize** the streamlines and \(C_p\) distribution.
- **Compare** our computed lift coefficient \(C_l\) with thin airfoil theory.

First, we import the necessary libraries:

## Discretization into Panels

We now discretize the airfoil surface into a series of panels (straight line segments in 2D).  
Each **panel** has attributes:
- starting and ending coordinates,
- midpoint (collocation point),
- length,
- orientation (angle of the normal vector relative to the x-axis).

We define a `Panel` class to store this information.  
For this notebook, we slightly adapt the `Panel` class from the cylinder example:  
in addition to the usual geometric data, we assign each panel a label (`upper` or `lower`) depending on whether it lies on the upper or lower surface of the airfoil.  
This is useful for plotting, allowing us to distinguish results above and below the chord line.

The orientation and ordering of panels is important, as it determines the direction of the normal vector and the application of boundary conditions.


```python
class Panel:
    """
    Contains information related to a panel.
    """
    def __init__(self, xa, ya, xb, yb):
        """
        Initializes the panel.
        
        Sets the end-points and calculates the center, length,
        and angle (with the x-axis) of the panel.
        Defines if the panel is on the lower or upper surface of the geometry.
        Initializes the source-sheet strength, tangential velocity,
        and pressure coefficient to zero.
        
        Parameters
        ----------
        xa: float
            x-coordinate of the first end-point.
        ya: float
            y-coordinate of the first end-point.
        xb: float
            x-coordinate of the second end-point.
        yb: float
            y-coordinate of the second end-point.
        """
        self.xa, self.ya = xa, ya
        self.xb, self.yb = xb, yb
        # Panel tangent angle (from xa,ya to xb,yb)
        self.theta = np.arctan2(yb - ya, xb - xa)
        # Outward normal angle
        self.beta = self.theta - np.pi/2
        # Panel center (collocation point)
        self.xc, self.yc = (xa + xb) / 2, (ya + yb) / 2
        self.length = np.sqrt((xb - xa)**2 + (yb - ya)**2)
        # Label panel as upper/lower by geometry, not by angle
        self.loc = 'upper' if self.yc > 0 else 'lower'
        # To be filled later:
        self.gamma = 0.0
        self.vt = 0.0
        self.cp = 0.0
```

## Non-Uniform Panel Distribution

For a circular cylinder, panel discretization is straightforward.  
For a general airfoil geometry (like NACA0012), we need a **non-uniform panel distribution** to capture the high curvature near the leading and trailing edges.

One common approach is:
- **Discretize a circle** whose diameter equals the airfoil’s chord, with the leading and trailing edges touching the circle.
- Use the **x-coordinates of the circle points** as the x-coordinates for the panel nodes.
- **Project these x-points onto the airfoil surface**: for each x-value, interpolate the corresponding y-value from the airfoil geometry.

This generates a node distribution that is **refined near the leading and trailing edges**, improving accuracy in regions of strong curvature.

The function `define_panels()` implements this:
- It returns an array of `Panel` objects, each containing all geometric and orientation information for a panel, given the desired number of panels and the body coordinates.

### Remarks on `define_panels()` implementation:
- Only the x-coordinates of the circle (`x_circle`) are needed; y-coordinates are found by interpolation.
- The circle is created with `N+1` points, so the first and last points coincide, ensuring a closed loop.
- NumPy arrays are extended to repeat the first value at the end, avoiding special cases at the endpoints.
- A while-loop finds two consecutive points on the airfoil so that their x-interval contains the current circle x-coordinate; linear interpolation then determines the corresponding y-value.

This approach ensures a smooth, non-uniform paneling that is denser where flow features are most critical.


```python
def define_panels(x, y, N=40):
    """
    Discretizes the geometry into panels using the 'cosine' method.
    
    Parameters
    ----------
    x: 1D array of floats
        x-coordinate of the points defining the geometry.
    y: 1D array of floats
        y-coordinate of the points defining the geometry.
    N: integer, optional
        Number of panels;
        default: 40.
    
    Returns
    -------
    panels: 1D Numpy array of Panel objects
        The discretization of the geometry into panels.
    """
    R = (x.max() - x.min()) / 2  # radius of the circle
    x_center = (x.max() + x.min()) / 2  # x-coord of the center
    # define x-coord of the circle points
    x_circle = x_center + R * np.cos(np.linspace(0.0, 2 * np.pi, N + 1))
    
    x_ends = np.copy(x_circle)  # projection of the x-coord on the surface
    y_ends = np.empty_like(x_ends)  # initialization of the y-coord Numpy array

    x, y = np.append(x, x[0]), np.append(y, y[0])  # extend arrays using numpy.append
    
    # computes the y-coordinate of end-points
    I = 0
    for i in range(N):
        while I < len(x) - 1:
            if (x[I] <= x_ends[i] <= x[I + 1]) or (x[I + 1] <= x_ends[i] <= x[I]):
                break
            else:
                I += 1
        a = (y[I + 1] - y[I]) / (x[I + 1] - x[I])
        b = y[I + 1] - a * x[I + 1]
        y_ends[i] = a * x_ends[i] + b
    y_ends[N] = y_ends[0]
    
    panels = np.empty(N, dtype=object)
    for i in range(N):
        panels[i] = Panel(x_ends[i], y_ends[i], x_ends[i + 1], y_ends[i + 1])
    
    return panels
```

Now we can use this function, calling it with a desired number of panels whenever we execute the cell below. We also plot the resulting geometry.

```python
N = 40                           # number of panels
panels = define_panels(x, y, N)  # discretizes of the geometry into panels

# plot the geometry and the panels
width = 10
plt.figure(figsize=(width, width))
plt.grid()
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.plot(x, y, color='k', linestyle='-', linewidth=2)
plt.plot(np.append([panel.xa for panel in panels], panels[0].xa),
            np.append([panel.ya for panel in panels], panels[0].ya),
            linestyle='-', linewidth=1, marker='o', markersize=6, color='#CD2305')
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 0.1)
plt.axis('equal')
```

## Freestream Conditions

We will immerse the NACA0012 airfoil in a uniform flow with velocity $U_\infty$ and angle of attack $\alpha $.  
For this example, we start with $ \alpha = 0 $.

Even though there is only one freestream in this problem, we define a `Freestream` class to:
- Make the code **cleaner and more readable**.
- Avoid naming conflicts with variables like `u_inf` and `alpha` elsewhere in the code.
- Easily pass all freestream conditions to functions by passing a single object, instead of multiple arguments.

This approach keeps the code modular and easier to extend for more complex simulations.


```python
class Freestream:
    """
    Freestream conditions.
    """
    def __init__(self, u_inf=1.0, alpha=0.0):
        """
        Sets the freestream speed and angle (with the x-axis).
        
        Parameters
        ----------
        u_inf: float, optional
            Freestream speed;
            default: 1.0.
        alpha: float, optional
            Angle of attack in degrees;
            default: 0.0.
        """
        self.u_inf = u_inf
        self.alpha = np.radians(alpha)  # degrees --> radians
```

```python
# define and creates the object freestream
u_inf = 1.0                            # freestream spee
alpha = 0.0                            # angle of attack (in degrees)
freestream = Freestream(u_inf, alpha)  # instantiation of the object freestream
```

# Flow Tangency Boundary Condition

Enforcing the **flow-tangency (no-penetration) boundary condition** at the control point (center) of each panel ensures the body becomes a dividing streamline for the potential flow. The accuracy improves as the number of panels increases.

For each panel $i$, we require:
$$
u_{n,i} = \frac{\partial \phi}{\partial n_i} (x_{c,i}, y_{c,i}) = 0
$$

In the **vortex panel method**, this becomes:
$$
0 = U_\infty \cos(\beta_i - \alpha)
  + \sum_{\substack{j=1\\j\neq i}}^{N_p} \frac{\gamma_j}{2\pi}
    \int_{0}^{L_j}
      \frac{\partial\theta_{ij}}{\partial n_i}\,ds_j
$$

Where:
- $(x_{c,i}, y_{c,i})$ = collocation point of panel $i$
- $\beta_i$ = outward normal angle of panel $i$
- $\gamma_j$ = vortex strength of panel $j$
- $x_j(s), y_j(s)$ = coordinates along panel $j$, parameterized by length $s$

With:
$$
\frac{\partial\theta_{ij}}{\partial n_i}
= \frac{
  -\,(y_{c,i} - y_j(s)) \cos\beta_i
  +\,(x_{c,i} - x_j(s)) \sin\beta_i
}{
  (x_{c,i} - x_j(s))^2 + (y_{c,i} - y_j(s))^2
}
$$

where
$$
\theta_{ij} = \arctan\left(\frac{y_{c,i} - y_j(s)}{x_{c,i} - x_j(s)}\right)
$$

---

# Tangential Derivative (for Pressure Coefficient)

After solving for the vortex strengths $\{\gamma_j\}$, the **surface-tangential velocity** at each panel $i$ is:
$$
u_{t,i} = \frac{\partial\phi}{\partial t_i}(x_{c,i}, y_{c,i})
= U_\infty \sin(\beta_i - \alpha)
+ \sum_{\substack{j=1\\j\neq i}}^{N_p} \frac{\gamma_j}{2\pi}
  \int_{0}^{L_j}
    \frac{\partial\theta_{ij}}{\partial t_i}\,ds_j
$$

Where:
$$
\frac{\partial\theta_{ij}}{\partial t_i}
= \frac{
  (y_{c,i} - y_j(s)) \sin\beta_i
  + (x_{c,i} - x_j(s)) \cos\beta_i
}{
  (x_{c,i} - x_j(s))^2 + (y_{c,i} - y_j(s))^2
}
$$

---

## Cartesian Derivatives (for Full Velocity Field)

To reconstruct the full velocity field $\mathbf{u} = (u, v)$ at arbitrary points:
$$
u = \frac{\partial\phi}{\partial x}, \quad
v = \frac{\partial\phi}{\partial y}
$$

using:
$$
\frac{\partial\theta_{ij}}{\partial x}
= \frac{-(y_c - y_j)}{(x_c - x_j)^2 + (y_c - y_j)^2},
$$
$$
\frac{\partial\theta_{ij}}{\partial y}
= \frac{x_c - x_j}{(x_c - x_j)^2 + (y_c - y_j)^2}
$$

---

*These formulas match the kernel used in your `integral` function for the panel matrix and tangential velocity computations.*


```python
def integral(x, y, panel, cos_beta_i, sin_beta_i):
    """
    Evaluates the contribution of a vortex‐panel at one target point.

    Uses the kernel ∂θij/∂ni:
        ∂θij/∂ni = [–(y_i − y_j)·cosβ_i + (x_i − x_j)·sinβ_i] 
                   / [ (x_i − x_j)^2 + (y_i − y_j)^2 ]

    Parameters
    ----------
    x : float
        x-coordinate of the collocation (target) point, x_i.
    y : float
        y-coordinate of the collocation (target) point, y_i.
    panel : Panel
        The source panel j (with end-point panel.xa, panel.ya, 
        orientation panel.beta, and length panel.length).
    cos_beta_i : float
        Cosine of the normal direction on panel i (i.e. cos β_i).
    sin_beta_i : float
        Sine of the normal direction on panel i (i.e. sin β_i).

    Returns
    -------
    float
        The line integral ∫ (∂θij/∂ni) ds_j over panel j.
    """
    def integrand(s):
        xj = panel.xa + np.cos(panel.theta) * s
        yj = panel.ya + np.sin(panel.theta) * s
        dx = x - xj
        dy = y - yj
        return (-dy * cos_beta_i + dx * sin_beta_i) / (dx*dx + dy*dy)
    value, _ = integrate.quad(integrand, 0.0, panel.length)
    return value
```

## Building the Linear System

We now build and solve the linear system of equations:

$$
[A][\gamma] = [b]
$$

where $[\gamma]$ is the vector of unknown vortex strengths for each panel.  
Each element of the matrix $[A]$ is computed by calling the `integral()` function with the appropriate projection: $\cos\beta_i$ and $\sin\beta_i$ (the components of the **outward normal** to panel $i$).

The right-hand side $[b]$ contains the **normal component of the freestream** at each control point.

Finally, we use `numpy.linalg.solve()` (or `numpy.linalg.lstsq()`, for least-squares if the matrix is not square) to solve the system and find the vortex strengths.

---

## Kutta Condition

To ensure a **finite velocity and smooth flow at the trailing edge** (the Kutta condition),  
we impose that the vortex strengths at the two panels meeting at the trailing edge satisfy

$$
\gamma_N + \gamma_1 = 0 \qquad \Longrightarrow \qquad \gamma_N = -\gamma_1
$$

(where $N$ is the total number of panels).

In practice, we enforce this by **replacing the last row** of the matrix system (normally the boundary condition at one panel) with the Kutta condition above, so the linear system remains square and solvable.

This guarantees that the circulation is chosen so that the flow leaves the trailing edge smoothly—just as in classical airfoil theory.


```python
def build_matrix(panels):
    """
    Builds the source matrix.
    
    Parameters
    ----------
    panels: 1D array of Panel object
        The source panels.
    
    Returns
    -------
    A: 2D Numpy array of floats
        The source matrix (NxN matrix; N is the number of panels).
    """
    N = len(panels)
    A = np.zeros((N, N))
    for i, p_i in enumerate(panels):
        cos_b = np.cos(p_i.beta)
        sin_b = np.sin(p_i.beta)
        for j, p_j in enumerate(panels):
            if i == j:
                A[i, j] = 0.0  # For vortex panels, self-induced normal velocity = 0
            else:
                A[i, j] = (1/(2*np.pi)) * integral(
                    p_i.xc, p_i.yc, p_j, cos_b, sin_b
                )
    # Kutta condition: γ1 + γN = 0 (enforce at last row)
    A[-1, :] = 0.0
    A[-1, 0] = 1.0
    A[-1, -1] = 1.0
    return A

def build_rhs(panels, freestream):
    """
    Builds the RHS of the linear system.
    
    Parameters
    ----------
    panels: 1D array of Panel objects
        The source panels.
    freestream: Freestream object
        The freestream conditions.
    
    Returns
    -------
    b: 1D Numpy array of floats
        RHS of the linear system.
    """    
    N = len(panels)
    b = np.zeros(N)
    for i, p in enumerate(panels):
        b[i] = -freestream.u_inf * np.cos(freestream.alpha - p.beta)
    b[-1] = 0.0  # Kutta condition row
    return b
```

```python
A = build_matrix(panels)           # compute the singularity matrix
b = build_rhs(panels, freestream)  # compute the freestream RHS
# solve the linear system
gamma = np.linalg.solve(A, b)

for i, panel in enumerate(panels):
    panel.gamma = gamma[i]
```

## Pressure Coefficient on Each Panel

After solving for the vortex strengths, we use **Bernoulli’s equation** to compute the pressure coefficient on each panel:

$$
C_{p_i} = 1 - \left(\frac{u_{t_i}}{U_\infty}\right)^2
$$

where $u_{t_i}$ is the **tangential velocity** at the midpoint of the $i$-th panel, calculated as:

$$
u_{t_i}
= U_\infty \sin(\beta_i - \alpha)
+ \sum_{\substack{j=1\\j\neq i}}^{N_p}
    \frac{\gamma_j}{2\pi}
    \int_{0}^{L_j}
      \frac{\partial\theta_{ij}}{\partial t_i}\,\mathrm{d}s_j
$$

where:
- $U_\infty$ is the freestream speed,
- $\beta_i$ is the tangent direction angle of panel $i$,
- $\alpha$ is the angle of attack,
- $\gamma_j$ is the vortex strength of panel $j$.

The **kernel** for the tangential projection is:

$$
\frac{\partial\theta_{ij}}{\partial t_i}
= 
\frac{
  (y_{c_i}-y_j(s))\,\sin\beta_i
  + (x_{c_i}-x_j(s))\,\cos\beta_i
}{
  (x_{c_i}-x_j(s))^2 + (y_{c_i}-y_j(s))^2
}
$$

This expression matches the one used in your code’s `integral()` function when evaluating the induced tangential velocity at each panel.



```python
def get_tangential_velocity(panels, freestream):
    """
    Computes the tangential velocity on the surface of the panels using the vortex panel method.

    Parameters
    ----------
    panels : list of Panel objects
        The vortex panels (must already have gamma assigned).
    freestream : Freestream object
        The freestream conditions (alpha in radians, u_inf).
    """
    for i, p_i in enumerate(panels):
        # Panel tangent direction: (cos(theta), sin(theta))
        tan_x = np.cos(p_i.theta)
        tan_y = np.sin(p_i.theta)
        vt = freestream.u_inf * np.cos(freestream.alpha - p_i.theta)  # Freestream tangent component

        for j, p_j in enumerate(panels):
            if i == j:
                continue
            vt += (1/(2*np.pi)) * p_j.gamma * integral(
                p_i.xc, p_i.yc, p_j,
                tan_x, tan_y
            )
        p_i.vt = vt
```

```python
# compute the tangential velocity at the center-point of each panel
get_tangential_velocity(panels, freestream)
```

```python
def get_pressure_coefficient(panels, freestream):
    """
    Computes the surface pressure coefficients on the panels.
    
    Parameters
    ---------
    panels: 1D array of Panel objects
        The vortex panels, each with a computed `.vt` (tangential velocity).
    freestream: Freestream object
        The freestream conditions.
    """
    for p in panels:
        p.cp = 1.0 - (p.vt / freestream.u_inf)**2
```

```python
# computes the surface pressure coefficients
get_pressure_coefficient(panels, freestream)
```

### Theoretical solution
There is a classical method to obtain the theoretical characteristics of airfoils, known as *Theodorsen's method*. It uses the Joukowski transformation but is able to deal with any airfoil by an additional transformation between a "near circle" and a circle. The method is hairy indeed! But the resulting values of pressure coefficient are provided for some airfoils in table form in the 1945 [NACA Report No.824](http://ntrs.nasa.gov/archive/nasa/casi.ntrs.nasa.gov/19930090976.pdf), available from the NASA web server (see p. 71). 

The values of $(u/U_{\infty})^2$ are given for several stations along the chord length. We transcribed them here, saving them into an array:

```python
voverVsquared=np.array([0.0, 0.64, 1.01, 1.241, 1.378, 1.402, 1.411, 1.411,
                           1.399, 1.378, 1.35, 1.288, 1.228, 1.166, 1.109, 1.044,
                           0.956, 0.906, 0.0])
print(voverVsquared)
xtheo=np.array([0.0, 0.5, 1.25, 2.5, 5.0, 7.5, 10.0, 15.0, 20.0, 25.0, 30.0,
                   40.0, 50.0, 60.0, 70.0, 80.0, 90.0, 95.0, 100.0])
xtheo /= 100
print(xtheo)
```

```python
# plot the surface pressure coefficient
plt.figure(figsize=(10, 6))
plt.grid()
plt.xlabel('x', fontsize=16)
plt.ylabel('$C_p$', fontsize=16)
plt.plot([panel.xc for panel in panels if panel.loc == 'upper'],
            [panel.cp for panel in panels if panel.loc == 'upper'],
            label='upper',
            color='r', linewidth=1, marker='x', markersize=8)
plt.plot([panel.xc for panel in panels if panel.loc == 'lower'],
            [panel.cp for panel in panels if panel.loc == 'lower'],
            label='lower',
            color='b', linewidth=0, marker='d', markersize=6)
plt.plot(xtheo, 1-voverVsquared,
            label='theoretical',
            color='k', linestyle='--',linewidth=2)
plt.legend(loc='best', prop={'size':14})
plt.xlim(-0.1, 1.1)
plt.ylim(1.0, -0.6)
plt.title('Number of panels : {}'.format(N));
```

```python
import numpy as np

def debug_panel_method(panels, freestream):
    """
    Compute various diagnostics to check physical consistency:
      1. Normal boundary condition residuals at control points.
      2. Tangential velocity at TE (stagnation point) check.
      3. Kutta condition residual: γ1 + γN.
      4. Total circulation and lift coefficient vs. theory.

    Parameters
    ----------
    panels : list of Panel
        Each Panel has attributes xc, yc, beta, length, gamma.
    freestream : Freestream
        With attributes u_inf, alpha (in radians), and chord c.

    Prints:
        - max|u_n| residual
        - |u_t(TE)| residual
        - |γ1 + γN|
        - computed Cl vs theoretical 2π α
    """
    # 1) Normal condition residuals
    u_n_res = []
    for p_i in panels:
        # freestream normal component (neglect sign convention)
        u_n = freestream.u_inf * np.cos(freestream.alpha - p_i.beta)
        # influence of vortices
        for p_j in panels:
            if p_i is p_j:
                continue
            u_n += p_j.gamma/(2*np.pi) * integral(
                p_i.xc, p_i.yc, p_j,
                np.cos(p_i.beta), np.sin(p_i.beta)
            )
        u_n_res.append(u_n)
    u_n_res = np.array(u_n_res)
    
    # 2) Tangential at TE
    te = panels[-1]
    u_t_te = freestream.u_inf * np.sin(freestream.alpha - te.beta)
    for p_j in panels:
        u_t_te += p_j.gamma/(2*np.pi) * integral(
            te.xb, te.yb, p_j,
            np.sin(te.beta), -np.cos(te.beta)
        )
    
    # 3) Kutta: gamma_1 + gamma_N
    gamma_sum = panels[0].gamma + panels[-1].gamma
    
    # 4) Total circulation and Cl
    total_circ = sum(p.gamma * p.length for p in panels)
    # chord: assume x runs from 0 to c
    c = max(max(p.xa, p.xb) for p in panels) - min(min(p.xa, p.xb) for p in panels)
    # lift coefficient Cl = circulation / (0.5 * Uinf * c)
    Cl = total_circ / (0.5 * freestream.u_inf * c)
    # theoretical thin‐airfoil Cl = 2π * alpha
    Cl_theory = 2 * np.pi * freestream.alpha
    
    print(f"Normal BC residuals: max|u_n| = {np.abs(u_n_res).max():.3e}")
    print(f"Tangential velocity at TE: u_t(TE) = {u_t_te:.3e}")
    print(f"Kutta residual γ1+γN = {gamma_sum:.3e}")
    print(f"Total circulation Γ = {total_circ:.3e}")
    print(f"Computed Cl = {Cl:.3f}, Theory 2π*α = {Cl_theory:.3f}")

debug_panel_method(panels, freestream)

```

as we can see, Kutta residual γ1+γN = 0.000e+00, Computed Cl = 0.000, Theory 2π*α = 0.000, Normal BC residuals are almost zero

### **Plotting Streamlines Around the Airfoil**

To visualize the flow field around the airfoil at a given angle of attack ($\alpha = 10^\circ$), we proceed as follows:

- The freestream is set to an angle of attack of $10^\circ$.
- The linear system for the vortex strengths on each panel is assembled and solved.
- A grid of points is created to cover the region around the airfoil.
- At each point on the grid, the total velocity is computed as the sum of the freestream velocity and the influence of all vortex panels.
- The result is plotted as streamlines, showing the flow pattern around the NACA 0012 airfoil at the specified angle of attack.

This visualization illustrates how the airfoil modifies the flow, and how the streamlines are deflected around the airfoil in accordance with the boundary conditions enforced by the vortex panel method.


```python
# Set freestream with angle of attack 10 degrees
alpha = 10
freestream = Freestream(u_inf=1.0, alpha=alpha)

# Solve the linear system for gamma (vortex strengths)
A = build_matrix(panels)
b = build_rhs(panels, freestream)
gamma = np.linalg.solve(A, b)
for i, panel in enumerate(panels):
    panel.gamma = gamma[i]

# Create mesh grid for flow visualization
X, Y = np.meshgrid(np.linspace(-0.1, 1.1, 200), np.linspace(-0.3, 0.3, 200))
u = freestream.u_inf * np.cos(freestream.alpha) * np.ones_like(X)
v = freestream.u_inf * np.sin(freestream.alpha) * np.ones_like(Y)

# Superimpose vortex influence from each panel
for panel in panels:
    dx = X - panel.xc
    dy = Y - panel.yc
    r2 = dx**2 + dy**2
    u += -panel.gamma * dy / (2 * np.pi * r2)
    v +=  panel.gamma * dx / (2 * np.pi * r2)

# Plot airfoil and streamlines
plt.figure(figsize=(12, 5))
plt.grid()
plt.plot(x, y, color='k', linewidth=2)
plt.streamplot(X, Y, u, v, color='k', density=2, arrowsize=1)
plt.fill([panel.xa for panel in panels], [panel.ya for panel in panels], 'k', lw=2, alpha=0.2)
plt.title('Streamlines over NACA 0012, $\\alpha=10^\\circ$')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.xlim(-0.1, 1.1)
plt.ylim(-0.3, 0.3)
plt.show()
```

### **Calculating $C_L$ as a Function of Angle of Attack**

To analyze how the lift coefficient $C_L$ varies with angle of attack $\alpha$, we solve the vortex panel method for a range of angles from $-7^\circ$ to $15^\circ$. For each value of $\alpha$:

- The freestream angle in the simulation is updated.
- The linear system for the vortex strengths is reassembled and solved.
- The total circulation around the airfoil is calculated, and used to determine the lift coefficient $C_L$ for that angle of attack.
- The theoretical lift coefficient from thin airfoil theory, $C_L = 2\pi\alpha$, is also computed for comparison.

The resulting lists, `cls` and `cls_theory`, contain the panel method and theoretical lift coefficients, respectively, for each angle of attack in `alphas_deg`. This allows us to compare the numerical results from the panel method to the classical theory across a range of angles.


```python
alphas_deg = np.linspace(-7, 15, 24)  # in degrees
cls = []
cls_theory = []

for alpha_deg in alphas_deg:
    alpha_rad = np.radians(alpha_deg)
    # Update freestream object
    freestream.alpha = alpha_rad
    
    # Recompute RHS and solve for gamma
    b = build_rhs(panels, freestream)
    A = build_matrix(panels)
    gamma = np.linalg.solve(A, b)
    for i, panel in enumerate(panels):
        panel.gamma = gamma[i]
    
    # Compute total circulation
    circ = sum(panel.gamma * panel.length for panel in panels)
    c = max(max(p.xa, p.xb) for p in panels) - min(min(p.xa, p.xb) for p in panels)
    cl = -circ / (0.5 * freestream.u_inf * c)
    cls.append(cl)
    
    # Theory
    cls_theory.append(2 * np.pi * alpha_rad)

alphas = alphas_deg  # for x-axis
```

```python
plt.figure(figsize=(8,6))
plt.plot(alphas, cls, 'o-', label='Vortex panel method')
plt.plot(alphas, cls_theory, '--', label='Thin airfoil theory (2π·α)')
plt.xlabel(r'$\alpha$ [deg]', fontsize=14)
plt.ylabel(r'$C_l$', fontsize=14)
plt.grid(True)
plt.legend(fontsize=13)
plt.title('Lift coefficient $C_l$ vs Angle of Attack', fontsize=15)
plt.show()
```

### **Sign of the Lift Coefficient ($C_L$)**

In aerodynamics, the **lift coefficient $C_L$ is expected to be positive** for a **positive angle of attack** ($\alpha$) on a symmetric airfoil such as the NACA0012. Physically, this means that when the airfoil is tilted upward into the flow, it generates lift in the upward direction.

For the vortex panel method to yield physically correct results, it is important that all angle conventions, panel orientations, and circulation directions are consistent with standard aerodynamic definitions. After checking these aspects, the computed lift coefficient $C_L$ should increase with increasing $\alpha$ and closely follow the theoretical prediction for a thin symmetric airfoil, $C_L = 2\pi\alpha$.


```python
panel_counts = [16, 32, 64, 128]
alpha_deg = 5  # Angle of attack, can be changed
alpha_rad = np.radians(alpha_deg)

# Load geometry from file
x, y = np.loadtxt('naca0012.dat', dtype=float, delimiter='\t', unpack=True)

def compute_cl(N):
    panels = define_panels(x, y, N)
    freestream = Freestream(u_inf=1.0, alpha=alpha_deg)
    A = build_matrix(panels)
    b = build_rhs(panels, freestream)
    gamma = np.linalg.solve(A, b)
    for i, panel in enumerate(panels):
        panel.gamma = gamma[i]
    # Total circulation
    circulation = sum(panel.gamma * panel.length for panel in panels)
    chord = max(max(panel.xa, panel.xb) for panel in panels) - min(min(panel.xa, panel.xb) for panel in panels)
    Cl = -circulation / (0.5 * freestream.u_inf * chord)
    return Cl

# Calculate Cl for each panel count
cl_values = [compute_cl(N) for N in panel_counts]
theory_cl = 2 * np.pi * alpha_rad  # Thin airfoil theory for symmetric foil

# Plot results
plt.figure(figsize=(8,5))
plt.plot(panel_counts, cl_values, 'o-', label='Vortex panel method')
plt.axhline(theory_cl, color='k', linestyle='--', label='Thin airfoil theory')
plt.xlabel('Number of panels')
plt.ylabel('$C_L$')
plt.title(f'Lift coefficient $C_L$ vs Number of panels ($\\alpha={alpha_deg}^\\circ$)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
